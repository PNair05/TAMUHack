```python
import csv

input_file = 'input.csv'  # Replace with your input file name
output_file = 'toyotads.csv'

with open("test.csv", 'r', newline='') as infile, open("toyotads.csv", 'w', newline='') as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)
    
    for row in reader:
        cleaned_row = [cell.strip('"') for cell in row]
        writer.writerow(cleaned_row)

print(f"Processed file saved as {output_file}")
```

```python
def remove_quotes_from_file(input_file, output_file):
    try:
        # Read the file
        with open(input_file, 'r') as file:
            lines = file.readlines()

        # Remove quotation marks
        updated_lines = [line.replace('"', '') for line in lines]

        # Write the updated lines to a new file
        with open(output_file, 'w') as file:
            file.writelines(updated_lines)

        print(f"Quotation marks removed. Updated file saved as '{output_file}'.")
    except FileNotFoundError:
        print(f"Error: The file '{input_file}' does not exist.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Replace 'input.txt' and 'output.txt' with your file names
remove_quotes_from_file('test.csv', 'output.csv')
```
